```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(new_data.drop(['V28','V29','V30','V31','V32','V33','Class'],axis=1))
```

```python
scaled_features=scaler.transform(new_data.drop(['V28','V29','V30','V31','V32','V33','Class'],axis=1))
```

```python
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features,columns=new_data.columns[:-7])
```

```python
df_feat.head(5)
```

```python
df_feat[['V12','V21']].head(10)
```

```python
X=df_feat
```

```python
y=new_data[['V28','V29','V30','V31','V32','V33','Class']]
```

```python
y.head(5)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from keras.utils import np_utils
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
classifier=Sequential()
```

```python
X_train.head(5)
```

```python
classifier.add(Dense(output_dim=19,init='uniform',activation='relu',input_dim=26))
classifier.add(Dense(output_dim=15,init='uniform',activation='relu',))
classifier.add(Dense(output_dim=11,init='uniform',activation='relu',))
classifier.add(Dense(7,activation='softmax'))
```

```python
classifier.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size=20, nb_epoch=800)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
eval=classifier.evaluate(X_test,y_test,verbose=0)
```

```python
eval[0]
```

```python
eval[1]     #accuracy
```

```python
predict_class = np.argmax(y_pred, axis=1)
```

```python
predict_class = predict_class.tolist()
```
